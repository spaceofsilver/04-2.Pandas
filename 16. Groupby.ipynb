## Q. 호텔 등급"별" 가격의 합
- **groupby**( )사용
    - df.groupby( ).집계함수
        - 집계 가능한(숫자) 모든컬럼
- 실행되는 객체는?
    - **DataFrameGrupBy**객체
        - 집계 관련 함수:
            - sum( )
            - mean( )
            - count( )
            - max( )
            - min( )
            - **agg( )**
                - 여러개의 집계함수 사용할 때


```python
df.groupby('grade')[['price']].sum()
```

```python
df.groupby('grade')[['price']].mean().round(2)
```

```python
df.groupby('grade')[['price']].max()
```

```python
df.groupby('grade')[['price']].min()
```

```python
df.groupby('grade')[['price']].count()
```

## agg( )

```python
mydf = df.groupby('grade')[['price']].agg( func= ['mean','sum','max'] )
mydf.columns=['가격 평균','가격 합','가격 최대']
mydf
```
