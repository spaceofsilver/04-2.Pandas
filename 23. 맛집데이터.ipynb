```python
from bs4 import BeautifulSoup
import urllib.request as REQ

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib

from datetime import datetime 
# 폰트 설정
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
m_url = 'https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=bs203'
response = REQ.urlopen(m_url)
soup = BeautifulSoup( response, 'html5lib' ) # html.parser / html5lib
# soup
```

```python
s='body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul'
ulTag = soup.select_one( s )
foodTour = [ ]
rest_url = [ ]
for li in ulTag.select('li'):
#     print(li)
    print('순위:', li.select_one('.numTop, .rankNum').string )
    print('상호:', li.select_one('.restName').string )
    print('업종:', li.select_one('.listType').string )
    print('위치:', li.select_one('.listArea').string )
    foodTour.append( {   '순위': li.select_one('.numTop, .rankNum').string,
                         '상호': li.select_one('.restName').string,
                         '업종':li.select_one('.listType').string,
                         '주소': li.select_one('.listArea').string 
                     })
    rest_url.append( li.select_one('.restName > a')['href'] ) 
menu_df = pd.DataFrame( foodTour )
menu_df.set_index( '순위' , inplace=True )
# menu_df
```

```python
baseurl = 'https://www.menupan.com/'
tel = [ ]
addr = [ ] 
for i in rest_url:
#     m_url = 'https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=bs203'
    response = REQ.urlopen(baseurl+i)
    soup = BeautifulSoup( response, 'html5lib' )
#     print( baseurl+i )
    print('전화번호', soup.select_one('.tel1').text)
    tel.append(soup.select_one('.tel1').text)
    print( '실주소', soup.select_one('.add2').text)
    addr.append( soup.select_one('.add2').text )
```

```python
menu_df['전화번호'] = tel
menu_df['상세주소'] = addr
# menu_df.drop( columns = '주소', inplace=True)
menu_df
```
